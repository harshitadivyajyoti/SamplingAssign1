```python
import numpy as np
import pandas as pd
np.random.seed(0)
df=pd.read_csv('Creditcard_data.csv')
df.head(5)
```

```python
df['Class'].value_counts()
```

oversampling minority class

```python
df_majority = df[df['Class'] == 0]
df_minority = df[df['Class'] == 1]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(
    df_minority,
    replace=True,
    n_samples=len(df_majority),
    random_state=0
)
df_balanced = pd.concat([df_majority, df_minority_upsampled])
df_balanced = df_balanced.sample(frac=1, random_state=0).reset_index(drop=True)
df_balanced['Class'].value_counts()
```

simple random sampling

```python
sample_size=5
random_sample=df_balanced.sample(n=sample_size,random_state=42)
print(random_sample)
```

systematic sampling

```python
import math
import random
n = len(df_balanced)
k = int(math.sqrt(n))
random.seed(42)
start=random.randint(0,k-1)
systematic_sample = df_balanced.iloc[start::k][:5] #selecting every kth row with random start
print(systematic_sample.head())
```

stratified sampling

```python
from sklearn.model_selection import train_test_split
stratified_sample, _ = train_test_split(df_balanced,  test_size=len(df_balanced)-5, stratify=df_balanced['Class'], random_state=42)
print(stratified_sample.head())
```

cluster sampling

```python
df_temp = df_balanced.copy()
df_temp['cluster_col'] = df_temp.index % 10
clusters_to_pick = random.sample(list(df_temp['cluster_col'].unique()), 5)
cluster_sample = pd.concat([
    df_temp[df_temp['cluster_col'] == c].sample(n=1, random_state=42)
    for c in clusters_to_pick
])
print(cluster_sample.head())
```

```python
X_cluster = cluster_sample.drop(['Class','cluster_col'], axis=1).copy()
y_cluster = cluster_sample['Class'].copy()
```

probability proportional to size sampling

```python
sample_size = 5
weights = df_balanced['Class'].value_counts(normalize=True)
row_weights = df_balanced['Class'].map(weights)
pps_sample = df_balanced.sample(n=sample_size, weights=row_weights, random_state=42)
print(pps_sample)
```

models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
samples = [
    (random_sample.drop('Class', axis=1), random_sample['Class']),
    (systematic_sample.drop('Class', axis=1), systematic_sample['Class']),
    (stratified_sample.drop('Class', axis=1), stratified_sample['Class']),
    (X_cluster, y_cluster),
    (pps_sample.drop('Class', axis=1), pps_sample['Class'])
]
sample_names = ['Random', 'Systematic', 'Stratified', 'Cluster', 'PPS']
```

```python
models = [
    LogisticRegression(max_iter=1000),
    DecisionTreeClassifier(),
    RandomForestClassifier(n_estimators=50),
    KNeighborsClassifier(),
    SVC()
]
model_names = ['LogReg', 'DecisionTree', 'RandomForest', 'KNN', 'SVM']
```

```python
accuracy_matrix = pd.DataFrame(index=sample_names, columns=model_names)
```

remaining as test set

```python
all_sample_indices = list(random_sample.index) + list(systematic_sample.index) + \
                     list(stratified_sample.index) + list(cluster_sample.index) + \
                     list(pps_sample.index)

X_test = df_balanced.drop(all_sample_indices, axis=0).drop('Class', axis=1).copy()
y_test = df_balanced.drop(all_sample_indices, axis=0)['Class'].copy()
```

```python
for i, (X_samp, y_samp) in enumerate(samples):
    for j, model in enumerate(models):
        model.fit(X_samp, y_samp)
        y_pred = model.predict(X_test)
        accuracy_matrix.iloc[i, j] = accuracy_score(y_test, y_pred)
```

```python
print("5x5 Accuracy Matrix:\n")
print(accuracy_matrix)
```
